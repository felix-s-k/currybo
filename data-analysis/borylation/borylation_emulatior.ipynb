```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

original_dataset_etoh = pd.read_csv('etoh_borylation_original.csv', index_col=0)
original_dataset_meoh = pd.read_csv('meoh_borylation_original.csv', index_col=0)

original_dataset_etoh["electrophile_id"] = original_dataset_etoh["electrophile_id"].replace(electrophiles_smiles)
original_dataset_etoh["ligand_name"] = original_dataset_etoh["ligand_name"].replace(ligands_smiles)

original_dataset_etoh = original_dataset_etoh.rename(columns={"electrophile_id": "Electrophile", "ligand_name": "ligand"})
original_dataset_etoh['yield'] = original_dataset_etoh['yield'] / 100

print(original_dataset_etoh.head())
```

```python
original_dataset_meoh["solvent"] = "CO"
original_dataset_etoh["solvent"] = "CCO"

original_dataset = pd.concat([original_dataset_meoh, original_dataset_etoh], ignore_index=True)
original_dataset = original_dataset[["Electrophile", "ligand", "solvent", "yield"]]
print(original_dataset.head(), len(original_dataset))

original_dataset.to_csv('borylation_cleaned.csv', index=False)
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import rdFingerprintGenerator
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import root_mean_squared_error, mean_absolute_error
import pickle
import torch


# Function to convert SMILES to Morgan Fingerprint
def smiles_to_fingerprint(smiles, radius=2, n_bits=1024):
    if smiles is np.nan:
        return torch.zeros(n_bits, dtype=torch.float64)
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return torch.zeros(n_bits, dtype=torch.float64)
    #fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=n_bits)
    mfpgen = rdFingerprintGenerator.GetMorganGenerator(radius=radius, fpSize=n_bits)
    fp = mfpgen.GetFingerprint(mol)
    return torch.tensor(fp, dtype=torch.float64)
```

```python
fingerprints = []
for idx, row in original_dataset.iterrows():
    ligand_fp = smiles_to_fingerprint(row['ligand'])
    solvent_fp = smiles_to_fingerprint(row['solvent'])
    electrophile_fp = smiles_to_fingerprint(row['Electrophile'])
    combined_fp = np.concatenate([ligand_fp, solvent_fp, electrophile_fp])
    fingerprints.append(combined_fp)

X = np.array(fingerprints)
y = original_dataset['yield'].values
print(X.shape)
model = RandomForestRegressor(n_estimators=1000, random_state=12)
model.fit(X, y)

# Evaluate the model
y_pred = model.predict(X)
rmse = root_mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
print(f'Root Mean Squared Error: {rmse}, Mean Absolute Error: {mae}')

# Save the trained model
with open('borylation.pkl', 'wb') as f:
    pickle.dump(model, f)
```
